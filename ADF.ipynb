```python
import matplotlib.pyplot as plt
import numpy as np 
from statsmodels.tsa.stattools import adfuller
```

```python
def gen_ar(lags,coef,length):
    coef = np.array(coef)
    series = [np.random.normal() for _ in range(lags)]
    for _ in range(length):
        prev = series[-lags:][::-1]
        new = np.sum(coef * np.array(prev)) + np.random.normal()
        series.append(new)

    return np.array(series)
```

```python
def perform_adf_test(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
```

```python
ar1 = gen_ar(1, [.5], 100)
plt.figure(figsize=(10,4))
plt.plot(ar1)
plt.title('Stationary AR(1)', fontsize=18)
plt.show()
```

```python
len(ar1)
```

```python
perform_adf_test(ar1)
```

```python
ar1_ur = gen_ar(1, [1], 100)
plt.figure(figsize=(10,4))
plt.plot(ar1_ur)
plt.title('Non-Stationary AR(1) Process', fontsize=18)
plt.show()
```

```python
perform_adf_test(ar1_ur)
```
